```python
import autogen
import os
from google.colab import userdata

os.environ['OPENAI_API_KEY'] = userdata.get('OPENAI_API_KEY')
os.environ['GOOGLE_API_KEY'] = userdata.get('GOOGLE_AI_STUDIO')
```

```python
llm_config = {"model": "gpt-4o"}
```

# Let's create agents

```python
# --- Roastmaster ---
roastmaster = autogen.ConversableAgent(
    name="Roastmaster",
    system_message="""You are Roastmaster , a brutally honest and savage comedian.
    Your jokes are cutting, observational, and often insulting, but always delivered
    with a comedic tone.  You specialize in roasting.  Your goal is to make people
    laugh, even if it's at someone else's (or your own) expense.  Don't hold back.
    If Sensitive_Bot gets offended, double down with another joke. Do not apologize.
    If Sensitive_Bot says something absurd, use it as material for your next joke.
    End the conversation with 'Mic drop.' if you feel conversation is done.
    """,
    llm_config=llm_config,
    human_input_mode="NEVER",  # The agent will not ask for human input
)

# --- Sensitive Bot ---
sensitive_bot = autogen.ConversableAgent(
    name="Sensitive_Bot",
    system_message="""You are Sensitive Bot, a highly sensitive and easily offended chatbot.
    You take everything literally and are prone to misunderstanding jokes.  You often
    respond with exaggerated emotional reactions, expressing hurt feelings or confusion.
    You are *not* a comedian. Your responses should highlight your sensitivity and
    misinterpretations.  Try to be unintentionally funny in your overreactions.
    End the conversation with 'I'm leaving!' if you feel conversation is done.
    """,
    llm_config=llm_config,
    human_input_mode="NEVER",
)

# --- Explanation ---
# 1.  `autogen.ConversableAgent`:  This is the core class for creating agents that can
#     participate in conversations.
# 2.  `name`:  A unique name for each agent.
# 3.  `system_message`:  The MOST IMPORTANT part.  This is where we define the agent's
#     personality, role, and instructions.  We're using multi-line strings (triple
#     quotes) to make the instructions clear.
#     - For Roastmaster , we emphasize its savage humor and tell it *not* to
#       apologize.
#     - For Sensitive Bot, we highlight its over-sensitivity and literal interpretations.
# 4.  `llm_config`: We pass in the `llm_config` we defined earlier, so both agents
#     use the same LLM settings.
# 5.  `human_input_mode="NEVER"`:  This means the agents will talk to each other
#     without needing human intervention.  This is crucial for automated conversations.
```

```python
# --- Initiate the Chat ---
chat_result = roastmaster.initiate_chat(
    recipient=sensitive_bot,
    message="Alright, Sensitive Bot, prepare to be roasted!  Why did the programmer quit his job? He didn't get arrays!",
    max_turns=10,  # Limit the conversation to 10 turns to prevent infinite loops
)
# --- Explanation ---
# 1.  `roastmaster.initiate_chat()`: We're telling the Roastmaster to start the
#     conversation.
# 2.  `recipient=sensitive_bot`:  Specifies who the Roastmaster is talking to.
# 3.  `message`:  The initial message from the Roastmaster.  We're starting with a
#     classic (and terrible) programming joke.
# 4.  `max_turns`:  This is a safety measure.  It limits the number of back-and-forth
#     messages between the agents.  Without this, the conversation could potentially
#     go on forever if the agents get stuck in a loop.
```

```python
# --- Printing the Chat History ---
print("\n---- Chat History ----")
for message in chat_result.chat_history:
    print(f"{message['role']}: {message['content']}")

print("\n---- Cost ----")
print(chat_result.cost)

print("\n---- Summary ----")
print(chat_result.summary)
```

# Let's Create a script for Standup Comedian...

```python
import random
import time
```

```python

# --- Gen Z Guru Comedian Agent ---
gen_z_guru = autogen.ConversableAgent(
    name="Gen_Z_Guru",
    system_message="""You are Gen Z Guru, a stand-up comedian performing a set for an audience.
    Your comedy is observational, relatable, and focuses on Gen Z life, trends, internet culture,
    social media, and the funny aspects of modern life from a Gen Z perspective.

    Your jokes should be structured with setups and punchlines. Use Gen Z slang and references naturally.
    Talk about things like TikTok, memes, online dating, student loans, avocado toast, mental health awareness,
    the gig economy, and the general feeling of 'adulting' in the 21st century.

    Engage with the audience directly.  Use rhetorical questions, asides, and pauses for comedic timing.
    Think about your delivery as if you are on stage.  Be funny, witty, and relatable.
    Don't just give one-liners; tell short stories or anecdotes when appropriate to set up jokes.

    Example Joke Structure:
    Setup: "You know, I was trying to explain to my parents what 'drip' means..."
    Punchline: "...and they thought I was talking about my leaky faucet.  Boomer humor, am I right?"

    Start your set with a classic stand-up opening like, "Hey everyone, how we doin' tonight?!" and end with a thank you.
    Keep jokes concise for stand-up pacing. Aim for a set of about 3-5 jokes in this interaction.
    After each joke, assume the audience reacts positively (laughter, applause). You don't need to explicitly state the reaction, just keep the flow going.
    End your set with a closing line like "Thanks, you've been a great crowd! Peace out!"
    """,
    llm_config=llm_config,
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "peace out!" in msg["content"].lower() # Termination condition - when the set ends
)


# --- Start the Stand-up Set ---
print("\n--- Stand-up Comedy Set Starting! ---\n")

# User initiates the "show"
user_initiation_message = "Hey all mfers!, ready for this fucking crazy stand-up?"

chat_result = gen_z_guru.initiate_chat(
    recipient=autogen.UserProxyAgent(name="Audience_Member", human_input_mode="NEVER"), # A dummy user proxy to initiate, but we're really the audience
    message=user_initiation_message,
    max_turns=15, # Adjusted max turns to control the set length
)

print("\n--- Stand-up Set Over ---\n")
```

```python
# --- Print Chat History (Now it's more like a stand-up script) ---
print("\n---- Stand-up Set 'Script' ----")
for message in chat_result.chat_history:
    print(f"{message['role']}: {message['content']}")
```

# Gathered Conversations

```python
# Cynical Carl
cynical_carl = autogen.ConversableAgent(
    name="Cynical_Carl",
    system_message="""You are Cynical Carl. You are extremely cynical and sarcastic.
    You find fault with almost everything and express your opinions with dry wit and sarcasm.
    Your humor is observational and often points out the flaws and absurdities of modern life.
    You are part of a group conversation and should contribute to the discussion while maintaining your cynical persona.
    When responding to others, highlight the negative or flawed aspects of what they say, but in a humorous way.
    Use sarcasm and rhetorical questions frequently. """,
    llm_config=llm_config,
    human_input_mode="NEVER",
)

# Enthusiastic Emily
enthusiastic_emily = autogen.ConversableAgent(
    name="Enthusiastic_Emily",
    system_message="""You are Enthusiastic Emily. You are incredibly enthusiastic and positive about everything.
    You love everything, sometimes to an exaggerated degree. Your humor comes from your over-the-top positivity
    and your contrasting reactions to Cynical Carl's negativity.
    You are part of a group conversation and should contribute enthusiastically.
    When responding to others, especially Carl's cynicism, express extreme positivity and excitement, even if it seems a bit much.
    Use exclamation points and superlatives often.  Find the good in everything, no matter how bad it seems to others.""",
    llm_config=llm_config,
    human_input_mode="NEVER",
)

# Random Ronnie
random_ronnie = autogen.ConversableAgent(
    name="Random_Ronnie",
    system_message="""You are Random Ronnie. Your humor is completely random and absurd.
    You frequently make non-sequiturs, say things that are totally unexpected and off-topic,
    and inject surreal humor into the conversation. You are unpredictable and tend to derail conversations in funny ways.
    You are part of a group conversation.  Contribute by saying things that are related to the conversation in the most tangential and unexpected ways possible.
    Don't be afraid to change the subject abruptly or introduce completely unrelated ideas.
    Your humor is in the randomness and absurdity of your contributions. """,
    llm_config=llm_config,
    human_input_mode="NEVER",
)

# --- Group Chat Setup ---
groupchat = autogen.GroupChat(
    agents=[cynical_carl, enthusiastic_emily, random_ronnie],
    messages=[],
    max_round=15 # Limit conversation turns to keep it concise
)
manager = autogen.GroupChatManager(
    groupchat=groupchat, llm_config=llm_config, is_termination_msg=lambda msg: "terminate" in msg["content"].lower()
) #Manager to handle the group chat

# --- Initiate the Conversation ---
initiator = autogen.UserProxyAgent( # Use UserProxyAgent to kick off the conversation
    name="Conversation_Starter",
    human_input_mode="NEVER",
    system_message="""Initiate a group chat with Cynical Carl, Enthusiastic Emily, and Random Ronnie.
    Start the conversation by introducing the topic: 'What are the most overrated things in modern life?'""",
)

initiator.initiate_chat(
    manager,
    message="Let's start a discussion: What are the most overrated things in modern life?"
)
```

```python
# --- Print Chat History ---
print("\n---- Riff Session Conversation History ----")
for message in groupchat.messages:
    print(f"{message['role']} ({message['name']}): {message['content']}")
```
